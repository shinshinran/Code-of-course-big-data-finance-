# 1.2021WHR
https://www.kaggle.com/code/ajaypalsinghlo/world-happiness-report-2021-world

```python
data= pd.read_csv(r'world-happiness-report-2021.csv')
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.iloc[:,6:12]
```

```python
d1=data.iloc[:,6:12]
```

```python
d1.head()
```

```python
d1.describe()
```

```python
# 进行标准化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
scaler.fit(d1)  
X2_FI = pd.DataFrame(scaler.transform(d1))  
X2_FI.head()  # 查看标准化后的数据列
```

```python
X2_FI.describe()
```

```python
# PCA
import matplotlib.pyplot as plt  
from sklearn.decomposition import PCA  
pca = PCA(n_components=X2_FI.shape[1]-1)  # 创建主成分分析对象，设定主成分数为样本特征数 - 1
reduced_x = pca.fit_transform(X2_FI)  # 基于已标准化数据创建主成分分析模型
covper = pca.explained_variance_  # 降维后的各主成分的方差值
covper = pd.DataFrame(np.round(covper, 3))  
#covper.to_excel('covper.xlsx')
```

```python
# 绘制碎石图
plt.plot(covper, 'bx--')
plt.xlabel('Component #n')
plt.ylabel('Variance')
plt.show()
```

```python
#输出载荷系数
pd.DataFrame(np.round(pca.components_, 3), columns = d1.columns)
```

```python
#输出累计方差贡献率
pd.DataFrame(data = np.round(pca.explained_variance_ratio_,3))
```

```python
used = []
corrs = []
for i, j in enumerate(df.corr().columns):
    for k in range(len(df.corr())):
        if ((df.corr().iloc[k, i] > 0.9) & 
            (j not in used) &
            (j != df.corr().index[k])):
            
            used.append(j)
            corrs.append((j, df.corr().index[k], 
                          np.round(df.corr().iloc[k, i], 2)))

            
corrsdf = pd.DataFrame([[i[0] for i in corrs],
                        [i[1] for i in corrs],
                        [i[2] for i in corrs]])

corrsdf = corrsdf.T.rename(columns = {0:'column',1:'row',2:'corr'})
corrsdf[:15]
```
